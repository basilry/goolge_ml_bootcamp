### 코세라 | 신경망 및 딥러닝 2주차 - 더 많은 파생예제(도함수 관련)

```python
import matplotlib.pyplot as plt
import numpy as np

# 함수 정의
def f(x):
    return 2*x + 3

# 그래프를 그릴 범위
x = np.linspace(-10, 10, 400)
y = f(x)

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='$y = 2x + 3$', color='blue')

# 그래프 제목과 축 레이블 설정
plt.title('Graph of $y = 2x + 3$')
plt.xlabel('$x$')
plt.ylabel('$y$')

# 범례 설정
plt.legend()

# 그리드 추가
plt.grid(True)

# 그래프 보이기
plt.show()
```
